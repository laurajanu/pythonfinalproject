# AVARIJŲ STATISTIKA

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('duomenys.json')
```

```python
# kokios yra avarijų rūšys ir kiek kartų jos pasikartojo?
```

```python
duomenys = df['rusis']

kiekis = duomenys.value_counts()

for value, count in kiekis.items():
    print(f'{value}: {count}')
```

```python
duomenys = df['rusis']

kiekis = duomenys.value_counts()

for value, count in kiekis.items():
    print(f'{value}: {count}')


plt.bar(kiekis.index, kiekis.values)
plt.xlabel('Rūšys')
plt.ylabel('Kiekis')
plt.title('Avarijų rūšys')

plt.xticks(rotation=90)
plt.show()
```

```python
# kokios rūšies avarijų pasitaiko dažniausiai?
```

```python
duomenys = df['rusis']

kiekis = duomenys.value_counts()

didziausi = kiekis.nlargest(3)

for value, count in didziausi.items():
    print(f'viena dažniausių rūšių: {value}, {count}')
```

```python
duomenys = df['rusis']
kiekis = duomenys.value_counts()
didziausi = kiekis.nlargest(3)


table_data = []
for value, count in didziausi.items():
    table_data.append([value, count])

headers = ['Rūšys', 'Kiekis']


max_col_widths = [max(len(str(row[i])) for row in table_data + [headers]) for i in range(len(headers))]

header_str = ' | '.join(f'{header:<{max_col_widths[i]}}' for i, header in enumerate(headers))
print(header_str)
print('-' * len(header_str))


for row in table_data:
    row_str = ' | '.join(f'{str(cell):<{max_col_widths[i]}}' for i, cell in enumerate(row))
    print(row_str)
```

```python
# kiek iš viso žuvusiųjų? Palyginus su vaikais?
```

```python
zuvusiu = df['zuvusiuSkaicius'].sum()
vaiku = df['zuvVaiku'].sum()


print('iš viso žuvusių:', zuvusiu)


print('iš viso žuvusių vaikų:', vaiku)


if zuvusiu > vaiku:
    print('daugiau žūsta suaugę.')
elif zuvusiu < vaiku:
    print('daugiau žūsta vaikai.')
else:
    print('žuvusiųjų skaičius toks pat.')
```

```python
zuvusiu = df['zuvusiuSkaicius'].sum()
vaiku = df['zuvVaiku'].sum()

print('Iš viso žuvusiųjų:', zuvusiu)
print('Iš jų žuvusiųjų yra vaikai:', vaiku)

categories = ['Žuvusieji', 'Žuvusieji vaikai']
values = [zuvusiu, vaiku]

plt.bar(categories, values)
plt.xlabel('Žuvusieji(suaugę ir vaikai))')
plt.ylabel('Kiekis')
plt.title('Žuvusieji')


plt.show()
```

```python
# Ar girti/neblaivūs vairuotojai dažniau sukelia avarijas?
```

```python
duomenys = df['neblaivusKaltininkai']

kiekis = duomenys.value_counts()

ne = kiekis.get('Ne', 0)
taip = kiekis.get('Taip', 0)
kita = len(duomenys) - ne - taip
isviso = len(duomenys)

print(f'iš viso avarijų: {isviso}')
print(f'neblaivūs vairuotojai = {ne}')
print(f'blaivūs vairuotojai = {taip}')
print(f'kita = {kita}')
```

```python
duomenys = df['apsvaigeKaltininkai']

kiekis = duomenys.value_counts()

ne = kiekis.get('Ne', 0)
taip = kiekis.get('Taip', 0)
kita = len(duomenys) - ne - taip
isviso = len(duomenys)

print(f'iš viso avarijų: {isviso}')
print(f'neapsvaigę vairuotojai = {ne}')
print(f'apsvaigę vairuotojai = {taip}')
print(f'kita = {kita}')
```

```python
duomenys = df['atsisakeTikrintisKaltininkai']

kiekis = duomenys.value_counts()

ne = kiekis.get('Ne', 0)
taip = kiekis.get('Taip', 0)
kita = len(duomenys) - ne - taip
isviso = len(duomenys)

print(f'iš viso avarijų: {isviso}')
print(f'sutiko tikrintis = {ne}')
print(f'nesutiko tikrintis = {taip}')
print(f'kita = {kita}')
```

```python
neblaivus = df['neblaivusKaltininkai'].eq('Ne').sum()
apsvaige = df['apsvaigeKaltininkai'].eq('Taip').sum()
nesutiko = df['atsisakeTikrintisKaltininkai'].eq('Taip').sum()

print('iš viso avarijų:', isviso)
print('neblaiviu vairuotoju iš viso:', neblaivus)
print('apsvaigusių vairuotojų iš viso:', apsvaige)
print('atsisakiusių tikrintis iš viso:', nesutiko)
```

```python
neblaivus = df['neblaivusKaltininkai'].eq('Ne').sum()
apsvaige = df['apsvaigeKaltininkai'].eq('Taip').sum()
nesutiko = df['atsisakeTikrintisKaltininkai'].eq('Taip').sum()

print('Iš viso avarijų:', isviso)
print('Neblaiviu vairuotoju iš viso:', neblaivus)
print('Apsvaigusių vairuotojų iš viso:', apsvaige)
print('Atsisakiusių tikrintis iš viso:', nesutiko)

categories = ['Neblaivūs', 'Apsvaigę', 'Atsisakę tikrintis', 'Iš viso avarijų']
values = [neblaivus, apsvaige, nesutiko, isviso]

plt.bar(categories, values)
plt.show()
```
